### Q1. Reverse the NumPy array
arr = np.array([1, 2, 3, 6, 4, 5])


```python
import numpy as np

arr = np.array([1, 2, 3, 6, 4, 5])
reversed_arr = arr[::-1]

print("Original:", arr)
print("Reversed:", reversed_arr)
```

### Q2. Compare two NumPy arrays
arr1 = [[1,2],[3,4]]
arr2 = [[1,2],[3,4]]


```python
arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[1,2],[3,4]])

print("Element-wise comparison:\n", arr1 == arr2)
print("Arrays equal:", np.array_equal(arr1, arr2))
```

### Q3. Find most frequent value and indices


```python
# array x
x = np.array([1,2,3,4,5,1,2,1,1,1])
values, counts = np.unique(x, return_counts=True)
most_freq = values[np.argmax(counts)]
indices = np.where(x == most_freq)

print("Array x most frequent:", most_freq)
print("Indices:", indices)

# array y
y = np.array([1,1,1,2,3,4,2,4,3,3])
values, counts = np.unique(y, return_counts=True)
most_freq = values[np.argmax(counts)]
indices = np.where(y == most_freq)

print("\nArray y most frequent:", most_freq)
print("Indices:", indices)
```

### Q4. Matrix operations
gfg = [[4,1,9],[12,3,1],[4,5,6]]


```python
gfg = np.array([[4,1,9],
                [12,3,1],
                [4,5,6]])

print("Matrix:\n", gfg)

# total sum
print("\nTotal sum:", np.sum(gfg))

# row-wise sum
print("Row-wise sum:", np.sum(gfg, axis=1))

# column-wise sum
print("Column-wise sum:", np.sum(gfg, axis=0))
```
